#Explorando as diferenças entre Regressão linear simples e Regressão linear múltipla

##Começando com a regressão linear SIMPLES
##Importando Libs e estruturando variáveis

```
import numpy as np

## Crie x contendo um array em duas dimensões com os valores de Horas_de_Estudo já registrados no passado
##Repressanta a variável independente
x = np.array([1,5,7,8,10,11,14,15,15,19]).reshape((-1,1))


#Crie y contendo um array com os valores históricos da Nota já registrados no passado (vide tabela acima)
y = np.array([53,74,59,43,56,84,96,69,84,83])
```

## Importando a biblioteca de regressão linear do sklearn e criando o modelo

```
from sklearn.linear_model import LinearRegression

# Crie o modelo de regressão linear, fazendo o fit de x e y.
modelo = LinearRegression().fit(x,y)
```

##Avalie o modelo. Qual seu score, intercept e slope?

```
print("Score: ", modelo.score(x,y))
print("Intercept: ", modelo.intercept_)
print("Coef: ", modelo.coef_)
```

##Testando

```
#7. Crie novo_x contendo um array as seguintes horas de estudo dos novos alunos: 6, 9, 12, 15, 16, 4
novo_x = np.array([6,9,12,15,16,4]).reshape((-1,1))

#Predict
previsao = modelo.predict(novo_x)
print(previsao)
```

##Regressão Linear Mútipla
##Importando libs

```
import numpy as np
from sklearn.linear_model import LinearRegression
```

##Criando dados

```
x = np.array([[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]])
y = np.array([4, 5, 20, 14, 32, 22, 38, 43])
```

##Criando modelo

```
model = LinearRegression().fit(x, y)

print('coefficient of determination:', model.score(x, y))
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

##Testando

```
novo_x = np.arange(10).reshape((-1, 2))

previsao_y = model.predict(novo_x)
print(previsao_y)
```
