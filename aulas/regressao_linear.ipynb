```
#Cria um dataframe carregando os dados da base no formato CSV
df = pd.read_csv('/content/drive/My Drive/Inteligência Artificial/UA 9/aula 18/Python/carros.csv', sep=',')

#Imprimi as 6 primeiras linhas da base de dados
print(df.head())
```

#Exibe uma descrição da base

```
df.describe()
# print(df.describe())
```

#Separando as características / colunas / features para o processamento

```
motores = df[["ENGINESIZE"]]
co2 = df[["CO2EMISSIONS"]]

print(motores.head())
```

#Separando os elementos para treinamento e teste

```
motores_train, motores_test, co2_train, co2_test = train_test_split(motores, co2, test_size=0.2, random_state=42)

print(type(motores_test))
print(motores_test.describe())
```

#Plot dados

```
plt.scatter(motores_train, co2_train, color='blue')
plt.xlabel('Motor')
plt.ylabel('Emissão de CO 2')
plt.show()
```

#Criando modelo de aprendizagem

```
#Cria um modelo do tipo regressão linear
modelo =  linear_model.LinearRegression()

#Treinar modelo usando dataset para encontrar os valores de A e B em: Y = A + B.X
modelo.fit(motores_train, co2_train)
```

#Exibir coeficientes A e B

```
print('(A) Intercepto: ', modelo.intercept_)
print('(B) Inclinação: ', modelo.coef_)
```

#Plot reta gerada pelo modelo

```
plt.scatter(motores_train, co2_train, color='blue')
plt.plot(motores_train, modelo.coef_[0][0] * motores_train + modelo.intercept_[0], 'r-')
plt.xlabel('Motor')
plt.ylabel('Emissão de CO 2')
plt.show()
```

#Testando o modelo

```
predicoes_co2 = modelo.predict(motores_test)
```

#Vamos exibir a nossa reta de regressão no dataset de teste


```
plt.scatter(motores_test, co2_test, color='green')
plt.plot(motores_test, modelo.coef_[0][0] * motores_test + modelo.intercept_[0], 'b-')
plt.xlabel('Motor')
plt.ylabel('Emissão de CO 2')
plt.show()
```

#Avaliando o modelo


```
#Métricas
np.sum((predicoes_co2 - co2_test)**2))
print("Erro Quadrático Médio (MSE): %.2f" % mean_squared_error(co2_test, predicoes_co2))
print("Erro Médio Absoluto (MAE): %.2f" % mean_absolute_error(co2_test, predicoes_co2))
print ("Raiz do Erro Quadrático Médio (RMSE): %.2f " % sqrt(mean_squared_error(co2_test, predicoes_co2)))
print("R2-score: %.2f" % r2_score(predicoes_co2, co2_test) )
```
